```python
#written by chatGPT 5.1

RESULTS_DIR = os.path.join(base_data_folder, "../results")

RUN_FOLDERS = {
    "Seurat_3534.1762447849": "full_run_ZengAWS_Seurat_3534.1762447849",
    "mixhvg_3534.1763196049": "full_run_ZengAWS_mixhvg_3534.1763196049",
    "Original 2009":     "full_run_ZengAWS.1698256033",
    "half_HVG": "full_run_ZengAWS_half_hvg_1767.1763697258"
}

TOP_HITS_FILENAME = "top_hits.0.95.csv"
ASYM_FILENAME = "top_hits_asymmetric.0.99.best_vs_next.0.6.filtered.csv"

COL1 = "Study_ID|Celltype_1"
COL2 = "Study_ID|Celltype_2"


def canonicalize_pair_columns(df, col1=COL1, col2=COL2):
    """
    Ensure that in each row, df[col1] is alphabetically <= df[col2].
    If not, swap the two values in that row.
    """
    c1 = df[col1].astype(str)
    c2 = df[col2].astype(str)

    mask = c1 > c2
    df.loc[mask, [col1, col2]] = df.loc[mask, [col2, col1]].values
    return df


def load_reciprocal_pairs(run_label, folder_name, save_sorted=True):
    """
    Load one top_hits.0.95.csv, filter to Reciprocal_top_hit, canonicalize pairs,
    sort, optionally save, and return a set of (col1, col2) pairs.
    """
    in_path = os.path.join(RESULTS_DIR, folder_name, TOP_HITS_FILENAME)
    df = pd.read_csv(in_path)

    # Filter for Reciprocal_top_hit
    df_rec = df[df["Match_type"] == "Reciprocal_top_hit"].copy()

    # Canonicalize ordering within each row: col1 <= col2
    df_rec = canonicalize_pair_columns(df_rec, COL1, COL2)

    # Build a set of (celltype1, celltype2) pairs from canonicalized columns
    pairs = set(zip(df_rec[COL1], df_rec[COL2]))

    print(f"{run_label}: {len(pairs)} reciprocal pairs (canonicalized)")
    return pairs


def load_asymmetric_pairs(save_sorted=True):
    """
    Load top_hits_asymmetric.0.99.best_vs_next.0.6.filtered.csv from
    full_run_ZengAWS.1698256033, canonicalize pairs, optionally save,
    and return a set of (col1, col2) pairs.

    Assumes the file already represents the subset of interest,
    so no Match_type filtering is done.
    """
    folder_name = RUN_FOLDERS["Original 2009"]
    in_path = os.path.join(RESULTS_DIR, folder_name, ASYM_FILENAME)
    df = pd.read_csv(in_path).copy()

    # Canonicalize ordering within each row
    df = canonicalize_pair_columns(df, COL1, COL2)

    pairs = set(zip(df[COL1], df[COL2]))
    print(f"Asymmetric file: {len(pairs)} pairs (canonicalized)")
    return pairs


```

```python
# Load reciprocal pairs for all runs
run_pairs = {}
for run_label, folder_name in RUN_FOLDERS.items():
    run_pairs[run_label] = load_reciprocal_pairs(run_label, folder_name)

print("\nPair counts per run:")
for run_label, pairs in run_pairs.items():
    print(f"  {run_label}: {len(pairs)} pairs")

# Intersection across all three runs (canonicalized pairs)
all_runs_intersection = set.intersection(*run_pairs.values())
print("\nIntersection across ALL three top_hits.0.95 runs:")
print(f"  Common reciprocal pairs: {len(all_runs_intersection)}")

# Pairwise intersections (among the three top_hits.0.95 runs)
print("\nPairwise intersections among top_hits.0.95 runs:")
for (r1, r2) in itertools.combinations(run_pairs.keys(), 2):
    inter = run_pairs[r1].intersection(run_pairs[r2])
    print(
        f"  {r1} ∩ {r2}: {len(inter)} "
        f"(of {len(run_pairs[r1])} and {len(run_pairs[r2])})"
    )

# Save the intersection across all three runs
intersection_df = pd.DataFrame(
    list(all_runs_intersection),
    columns=[COL1, COL2],
).sort_values(by=[COL1, COL2])

out_intersection_path = os.path.join(
    RESULTS_DIR,
    "reciprocal_pairs_intersection_all_three_runs.canonical.csv",
)
intersection_df.to_csv(out_intersection_path, index=False)
print(f"\nSaved canonical intersection across all three runs to:\n  {out_intersection_path}")

# ---- Asymmetric file comparison ----
asym_pairs = load_asymmetric_pairs(save_sorted=True)

# Compare asymmetric file to each run
print("\nAsymmetric file intersections with each top_hits.0.95 run:")
for run_label, pairs in run_pairs.items():
    inter = asym_pairs.intersection(pairs)
    print(
        f"  Asymmetric ∩ {run_label}: {len(inter)} "
        f"(of {len(asym_pairs)} and {len(pairs)})"
    )

# Compare asymmetric file to intersection across all three runs
inter_asym_all_three = asym_pairs.intersection(all_runs_intersection)
print(
    "\nAsymmetric file ∩ (intersection of all three top_hits.0.95 runs): "
    f"{len(inter_asym_all_three)} "
    f"(of {len(asym_pairs)} asymmetric pairs and {len(all_runs_intersection)} common reciprocal pairs)"
)

# Optionally save that intersection too
inter_asym_all_three_df = pd.DataFrame(
    list(inter_asym_all_three),
    columns=[COL1, COL2],
).sort_values(by=[COL1, COL2])

out_asym_intersection_path = os.path.join(
    RESULTS_DIR,
    "asymmetric_vs_all_three_top_hits_intersection.canonical.csv",
)
inter_asym_all_three_df.to_csv(out_asym_intersection_path, index=False)
print(f"Saved asymmetric vs all-three intersection to:\n  {out_asym_intersection_path}")

```
