```python
base_data_folder = "/vault/lfrench/mouse_brain_cluster_replicability/data/"
```

```python
#from braincelldata.org
adata_merged = sc.read_h5ad(base_data_folder + '/whole_mouse_brain/processed/merged_Zeng_AWS.Oct2023.h5ad', backed=True)
```

```python
n_genes = adata_merged.n_vars

rng = np.random.default_rng(42)  # set seed for reproducibility
idx = np.sort(rng.choice(n_genes, size=n_genes // 2, replace=False))

# Subset while still backed, then load that subset into memory
adata_merged = adata_merged[:, idx]

```
